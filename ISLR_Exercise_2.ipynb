```python
# 9
auto_df = pd.read_csv("/content/Auto.csv")
```

```python
auto_df.head()
```

```python
auto_df.info()
```

```python
#checking for null values 
auto_df.isnull().any(axis = 1).sum()==0
```

```python
# droping missing values which had '?'
auto_df=auto_df.drop(auto_df[auto_df.values =='?'].index)
```

```python
auto_df[auto_df.values =='?']
```

```python
# (a):- Qualitative data:-  origin, name
# Quantitative data = mpg,cylinders,displacement, horsepower, weight, year
datatypes = {'quant': ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration'],
             'qual': ['origin', 'name']}
auto_quant = auto_df[datatypes['quant']].astype(np.float_)
```

```python
# 9(b,c)
pd.DataFrame({'range':auto_quant.max() - auto_quant.mean(),
              'Mean': auto_quant.mean(),
              'STD':auto_quant.std()})
```

```python
#9(d) 
auto_quant_d = auto_quant.drop(auto_quant.index[10:85])
pd.DataFrame({'Range':auto_quant_d.max()-auto_quant_d.min(),
              'Mean':auto_quant_d.mean(),
              'STD':auto_quant_d.std()})
```

```python
#9(e)
sns.pairplot(auto_df)
```
